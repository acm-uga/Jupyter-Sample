## Climate API visualization sample
In this sample we are requesting data from weatherstem's climate API for athens GA. We can pull a lot of cool information from their API and down below we can even pull a live image from one of the weather stations.

```python
import requests
import json
api_key = "vau6mwcs"
api_url = "https://api.weatherstem.com/api"
data = {"api_key":api_key,"stations":["uga@athensclarke.weatherstem.com"]}
data = json.dumps(data)
requests = requests.post(api_url,data=data)
data = requests.json()
print(data)
def returnImg(data):
    return data[0]['station']['cameras'][0]['image']
def returnLat(data):
    return data[0]['station']['lat']
def returnLong(data):
    return data[0]['station']['lon']
def returnHygrometer(data):
    return data[0]['record']['readings'][4]['value']
def returnSolarRadiation(data):
    return data[0]['record']['readings'][7]['value']
def returnTemp(data):
    return data[0]['record']['readings'][2]['value']
```

## Image of Athens
Using the code above we were able to pull an image from athens clarke county weather station using the API

```python
from IPython import display 
display.Image(returnImg(data))
```

## Current Climate Information


```python
from IPython.display import display, Markdown, Latex
display(Markdown("### Latitude: " + returnLat(data)))
display(Markdown("### Longitude: " + returnLong(data)))
display(Markdown("### Climate: "))
display(Markdown(returnHygrometer(data) + "% Humidity"))
display(Markdown(returnSolarRadiation(data) + " Watts per square meter of Solar Radiation"))
display(Markdown(returnTemp(data) + " Degrees Fahrenheit"))


```
